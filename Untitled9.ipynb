```python
%tensorflow_version 2.x  # this line is not required unless you are in a notebook
#import tensorflow and keras
import tensorflow as tf
from tensorflow import keras

#import helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=tf.keras.datasets.fashion_mnist                                          #load dataset from keras
(train_features,train_labels),(test_features,test_labels)=dataset.load_data()    #split dataset into training and testing

#take a look at data shape
print("Training Features Shape:", train_features.shape)
print("Training Labels Shape:", train_labels.shape)
print("Testing Features Shape:", test_features.shape)
print("Testing Labels Shape:", test_labels.shape)


print(train_features[0,12,12])   #look at 1 pixel (range from 0 to 255 where 0 states white and 255 states black)

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

#look at some images
plt.figure()
plt.imshow(train_features[3])
plt.colorbar()
plt.grid(True)
plt.show()

#pre-process data
train_features=train_features/255.0 # make training data values between 0 and 1 since neurons have an initial random weigths of range 0 to 1 so avoid missmatch
test_features=test_features/255.0   # make testing data values between 0 and 1 since neurons have an initial random weigths of range 0 to 1 so avoid missmatch

```

```python
model=keras.Sequential([
                        keras.layers.Flatten(input_shape=(28,28)),    #input layer
                        keras.layers.Dense(128,activation='relu'),    #hidden layer
                        keras.layers.Dense(10,activation='softmax')   #output layer
                        ])


model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy']) #compile model

model.fit(train_features,train_labels,epochs=1)#train model

test_loss,test_acc=model.evaluate(test_features,test_labels,verbose=1) #evaluate model
print('\nTest accuracy:', test_acc)

```

```python
predictions=model.predict(test_features)

print(predictions[0])
print(class_names[np.argmax(predictions[45])])
plt.figure()
plt.imshow(test_features[45])
plt.colorbar()
plt.grid(True)
plt.show()
print(test_labels[0])

```
